```python
import pandas as pd
import matplotlib.pyplot as plt

files = ['DDQN_lunar_main.csv', 'double_PER_lunar.csv', 'duelling_double_PER_lunar.csv', 'Vanilla_lunar.csv']
legend = ['Double DQN', 'Double DQN + Priority Experience Replay', 'Double DQN + Priority Experience Replay + Duelling', 'Vanilla DQN' ]

plt.figure(figsize=(20, 10))

def moving_average(data, window_size):
    """Calculate the moving average using a simple rolling mean."""
    return data.rolling(window=window_size, min_periods=1).mean()

for index, file in enumerate(files):
    data = pd.read_csv(file, header=None, names=['Episode Number','Reward','Epsilon','Avg. Loss'])
    moving_avg = moving_average(data['Reward'], window_size=50)  # Adjust window size as needed
    plt.plot(data['Episode Number'], moving_avg, label=f"{legend[index]}", linewidth=2)

plt.title('Moving Average Reward per Episode for Different RL Algorithms')
plt.xlabel('Episode Number')
plt.ylabel('Reward')

# Show legend
plt.legend()

# Show the plot
plt.show()
```
